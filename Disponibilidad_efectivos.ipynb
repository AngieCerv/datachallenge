```python
##cantidad en cas
def total_casetas(cas, pieces):
    total_cas = cas * pieces
    return total_cas
```

```python
df['caseta1'] = total_casetas(df['CAS_1'],df['PIEZAS_CAS_1'])
```

```python
df['caseta2'] = total_casetas(df['CAS_2'],df['PIEZAS_CAS_2'])
df['caseta3'] = total_casetas(df['CAS_3'],df['PIEZAS_CAS_3'])
df['caseta4'] = total_casetas(df['CAS_4'],df['PIEZAS_CAS_4'])
```

```python
df
```

```python
##suma
def sum_datos(cas1,cas2,cas3,cas4):
    suma = cas1+cas2+cas3+cas4
    return suma
```

```python
df['total']=sum_datos(df['caseta1'], df['caseta2'], df['caseta3'], df['caseta4'])
```

```python
##saldo total
df
```

```python
##pocentajes 25%
def percentage_criticality_c(min):
    crity_c = min * 0.25
    return crity_c
```

```python
##pocentajes 50%
def percentage_criticality_b(min):
    crity_b = min * 0.5
    return crity_b
```

```python
##pocentajes 75%
def percentage_criticality_a(min):
    crity_a = min * 0.75
    return crity_a
```

```python
df['crity_c'] = percentage_criticality_c(df['MINIMO'])
df['crity_b'] = percentage_criticality_b(df['MINIMO'])
df['crity_a'] = percentage_criticality_a(df['MINIMO'])
```

```python
df
```

```python
req = df[['CR','ATM','DIVISION','MINIMO','total','crity_c','crity_b','crity_a']].copy()
```

```python
req
```

```python
req['target'] = ''
req.loc[req['total'] > req['MINIMO'], 'target'] = 'OK'
req.loc[req['total']<=req['crity_a'], 'target'] = 'A'
req.loc[req['total']<=req['crity_b'], 'target'] = 'B'
req.loc[req['total']<=req['crity_c'], 'target'] = 'C'
req.loc[req['total'] == 0, 'target'] = 'Sin efectivo'
```

```python
req
```

```python
(req.isnull().sum()/len(req)).sort_values(ascending=False)
```

```python
req['crity_a']=req['crity_a'].fillna(req['crity_a'].mean())
```
